```python
import pygame
import config_c4
from c4_functions import (
    draw_circle, show_game_over 
)

def main():
    pygame.init()
    screen = pygame.display.set_mode((config_c4.WIDTH, config_c4.HEIGHT))
    pygame.display.set_caption("Example Mouse Click")
    clock = pygame.time.Clock()

    turn = 1

    player_x = False
    player_y = False

    running = True
    screen.fill(config_c4.BG_COLOR)
    while running:
        keys = pygame.key.get_pressed()
        for event in pygame.event.get():
            if keys[pygame.K_ESCAPE]:
                show_game_over(screen)
                running = False
                   
            if event.type == pygame.MOUSEBUTTONDOWN:
                if turn % 2 != 0:
                    mouse_pos = event.pos  # Get the mouse position (x, y)
                    player_x, player_y = mouse_pos
                    draw_circle(screen, player_x, player_y, 1)
                    turn += 1
                else:
                    mouse_pos = event.pos  # Get the mouse position (x, y)
                    player_x, player_y = mouse_pos
                    draw_circle(screen, player_x, player_y, 2)     
                    turn += 1       

        pygame.display.flip()

    pygame.quit()

if __name__ == "__main__":
    main()
```

```python
import pygame
import config
from adventure_functions import (
    draw_room, is_blocked, draw_obstacle, draw_player, load_player_image1, load_player_image2, show_game_over 
)

def main():
    pygame.init()
    screen = pygame.display.set_mode((config.WIDTH, config.HEIGHT))
    pygame.display.set_caption("3x3 Adventure Grid")
    clock = pygame.time.Clock()

    # Load the sprite ONCE
    load_player_image1()
    load_player_image2()
    turn = 0

    player_x = config.WIDTH // 2
    player_y = config.HEIGHT // 2

    running = True
    while running:
        dt = clock.tick(config.FPS) / 1000.0

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False
                show_game_over(screen)

        # Input
        keys= pygame.key.get_pressed()
        for even in pygame.event.get():
            if keys[pygame.K_ESCAPE]:
                running = False
                show_game_over(screen)
        keys = pygame.key.get_pressed()
        dx = dy = 0
        if keys[pygame.K_LEFT] or keys[pygame.K_a]: dx = -1
        if keys[pygame.K_RIGHT] or keys[pygame.K_d]: dx = 1
        if keys[pygame.K_UP] or keys[pygame.K_w]: dy = -1
        if keys[pygame.K_DOWN] or keys[pygame.K_s]: dy = 1
        if dx != 0 and dy != 0:
            dx *= 0.7071
            dy *= 0.7071

        new_x = player_x + dx * config.PLAYER_SPEED * dt
        new_y = player_y + dy * config.PLAYER_SPEED * dt

        # Movement with collision
        if not is_blocked(new_x, new_y):
            player_x, player_y = new_x, new_y

        # Draw everything
        screen.fill(config.BG_COLOR)

        # Rooms/walls first
        for row in range(config.GRID_ROWS):
            for col in range(config.GRID_COLS):
                draw_room(screen, row, col)

        # Any extra decor/obstacles
        draw_obstacle(screen)

        # Then the player
        draw_player(screen, player_x, player_y)

        pygame.display.flip()

    pygame.quit()

if __name__ == "__main__":
    main()
```

```python
import pygame

# Initialize Pygame
pygame.init()

# Set up the display
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Two Images")

# Load the images
# Replace 'image1.png' and 'image2.png' with your actual image file paths
image1 = pygame.image.load('/icons8-filled-circle-50.png').convert_alpha()  # .convert_alpha() for transparent images
image2 = pygame.image.load('icons8-purple-circle-48.png').convert_alpha()

# Get the rects for positioning
image1_rect = image1.get_rect()
image2_rect = image2.get_rect()

# Set initial positions for the images
image1_rect.topleft = (50, 50)  # Top-left corner at (50, 50)
image2_rect.center = (screen_width // 2, screen_height // 2)  # Center of the screen

# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the screen
    screen.fill((255, 255, 255))  # Fill with white background

    # Blit (draw) the images onto the screen
    screen.blit(image1, image1_rect)
    screen.blit(image2, image2_rect)

    # Update the display
    pygame.display.flip()  # or pygame.display.update()

# Quit Pygame
pygame.quit()
```
