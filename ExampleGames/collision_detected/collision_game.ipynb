```python
import pygame
import config_collision as config
from collision_functions import (draw_player, make_box, is_blocked, load_goose,
    show_game_over 
)

def main():
    pygame.init()
    screen = pygame.display.set_mode((config.WIDTH, config.HEIGHT))
    pygame.display.set_caption("What is this goose doing?")
    clock = pygame.time.Clock()
    #tmxdata = load_pygame('goosemap.tmx')

    # Load the sprite ONCE
    load_goose("R")

    player_x = config.WIDTH // 2
    player_y = config.HEIGHT // 2
    # define a rectangle for the player
    player_rect = pygame.Rect(player_x, player_y, config.PLAYER_RADIUS*2, config.PLAYER_RADIUS*2)
    counter = 1
    leg_step_wait = 30

    running = True
    while running:
        dt = clock.tick(config.FPS) / 1000.0
        # Update ball position
        keys = pygame.key.get_pressed()
        for event in pygame.event.get():
            if keys[pygame.K_ESCAPE]:
                show_game_over(screen)
                running = False

        # non-ESP Input
        dx = dy = 0
        if keys[pygame.K_LEFT] or keys[pygame.K_a]: 
            dx =- 1
            if counter % 2 == 0:
                load_goose("L")
                pygame.time.wait(leg_step_wait)
                counter += 1
            else:
                load_goose("LS")
                pygame.time.wait(leg_step_wait)
                counter += 1
        
        # animate the goose
        if keys[pygame.K_RIGHT] or keys[pygame.K_d]: 
            dx =+ 1
            if counter % 2 == 0:
                load_goose("R")
                pygame.time.wait(leg_step_wait)
                counter += 1
            else:
                load_goose("RS")
                pygame.time.wait(leg_step_wait)
                counter += 1
        if keys[pygame.K_UP] or keys[pygame.K_w]: dy = -1
        if keys[pygame.K_DOWN] or keys[pygame.K_s]: dy = 1
        if dx != 0 and dy != 0:
            dx *= 0.7071
            dy *= 0.7071
        new_x = player_x + dx * config.PLAYER_SPEED * dt
        new_y = player_y + dy * config.PLAYER_SPEED * dt
        
        # move player rectangle
        player_rect = pygame.Rect(new_x, new_y, config.PLAYER_RADIUS*2, config.PLAYER_RADIUS*2)
        # make a box they can collide with
        #a_box = pygame.Rect(new_x, new_y, config.WIDTH // 2, config.HEIGHT // 2)
        
        # Movement with collision
        if not is_blocked(new_x, new_y):
            player_x, player_y = new_x, new_y

        # Draw everything
        screen.fill(config.BG_COLOR)
        # Draw the player and the rectangle they can collide with
        
        draw_player(screen, player_x, player_y)
        
        a_box = make_box(screen, "simple_box")
        pygame.draw.rect(screen, config.BOX_COLOR, a_box)

        if player_rect.colliderect(a_box):
            show_game_over(screen)

        # pygame.draw.rect(screen, config.BOX_COLOR, a_box)
        pygame.display.flip()

    pygame.quit()

if __name__ == "__main__":
    main()
```
